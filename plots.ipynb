```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 1 with a small interval
x = np.linspace(0, 1, 100)

# Calculate y values for the corresponding x values
y = np.sqrt(x)

# Create a square figure and axis
fig, ax = plt.subplots(figsize=(5, 5))

# Plot the line graph
ax.plot(x, y)

# Add labels and title
#ax.set_xlabel('x')
#ax.set_ylabel('y')
#ax.set_title('y = sqrt(x)')

# Set aspect ratio to 'equal' for a square plot
ax.set_aspect('equal')

# Show the plot
#plt.show()

plt.savefig('plots/sqrt.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 1 with a small interval
x = np.linspace(0, 1, 100)

# Calculate y values for the corresponding x values
func = lambda v: max(.5, v)
y = [func(x_i) for x_i in x]

# Create a square figure and axis
fig, ax = plt.subplots(figsize=(5, 5))

# Plot the line graph
ax.plot(x, y)
ax.set_ylim([-0.05, 1.05])

# Add labels and title
#ax.set_xlabel('x')
#ax.set_ylabel('y')
#ax.set_title('y = sqrt(x)')

# Set aspect ratio to 'equal' for a square plot
ax.set_aspect('equal')

# Show the plot
#plt.show()

plt.savefig('plots/linear_compensated.png')
```
